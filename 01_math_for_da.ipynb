```python
# Скачивание картинки.
from PIL import Image, ImageOps
import requests

# Ссылка на картинку.
url = "https://code.s3.yandex.net/math-adult/explanations/linal/1/svd/frog.jpeg"

# Скачивание и сохранение картинки в переменную.
im = ImageOps.grayscale(Image.open(requests.get(url, stream=True).raw))
# Превращение картинки в NumPy-массив.
frog = np.array(im).astype(float)


def show_image(img):
	plt.figure(figsize=(12, 8))
	plt.imshow(img, cmap='gray')
	plt.xticks([])
	plt.yticks([])

show_image(frog)
```

```python
# выведем размеры изображения в пикселях
print(frog.shape)
```

```python
# получим матрицы для спектрального разложения рисунка
U, s, VT = np.linalg.svd(frog)
S = np.diag(s)
```

```python
# выведем размеры матриц для SVD оригинального рисунка
print('Матрица левых сингулярных векторов', U.shape)
print('Вектор сингулярных значений', S.shape)
print('Матрица правых сингулярных векторов транспонированная', VT.shape)
print()
```

```python
# вычислим размер исходного изображения
size = round(((frog.shape[0] * frog.shape[1]) * 8) / (1024 * 1024), 4)
print(f'Оригинальный размер изображения {size} Mб')
print()
```

```python
# зададим начальное значение параметра компонент изображения
k = 289

# зададим целевое значение веса картинки в мегабайтах
size_target = 1.2
```

```python
# вычислим суммарный размер трех матриц разложения
# при удалении наименее важного признака (первый шаг)

# обновляем матрицы
U = U[:, 0:k]
S = S[0:k, 0:k]
VT = VT[0:k, :]

# cчитаем размер матриц после удаления наименее важного признака
size = (
    (U.shape[0] * U.shape[1] +
    S.shape[0] +
    VT.shape[0] * VT.shape[1]) * 8
) / (1024 * 1024)

# обновляем значение параметра компонент изображения
k -= 1
```

```python
print(k, size)
```

```python
# создаем основной цикл поиска оптимального размера усеченного изображения
while (size > size_target):
    U = U[:, 0:k]
    S = S[0:k, 0:k]
    VT = VT[0:k, :]
    size = (
        (U.shape[0] * U.shape[1] +
        S.shape[0] +
        VT.shape[0] * VT.shape[1]) * 8
) / (1024 * 1024)
    k -= 1

# определим итоговое значение парамтера компонент изображения
k += 1

print(f'Число компонент матричного разложения {k}')
print(f'Размер усеченной (сжатой) картинки картинки {size} Мб')
```

```python
# получим матрицу усеченного изображения
frog_trunсated = U[:, 0:k] @ S[0:k, 0:k] @ VT[0:k, :]

# # выведем усеченное изображение
show_image(frog_trunсated)
```

```python
preprocessed_texts = np.array(['игра крутой большой онлайн',
                                 'огурец отстой не вкусный',
                                 'огурчик супер',
                                 'обычный огурчик понравиться соленый',
                                 'игра ужасный геймплей донат',
                                 'лопата шикарный взять на дача',
                                 'вкусный как на дача',
                                 'дрель мощь просверливать весь стена',
                                 'огурец понравиться свежий',
                                 'вкусный немного горьковатый'])

# Строим матрицу встречаемости слов.
count_model = CountVectorizer()
X = count_model.fit_transform(preprocessed_texts).toarray().astype(float).T
words = count_model.get_feature_names_out()
```

```python
# с помощью svd получим матрицы для матричного разложения
# матрицы встречаемости слов
U, s, VT = np.linalg.svd(X, full_matrices = False)
```

```python
# задаем количество используемых сингулярных векторов / топиков
k = 10
```

```python
# строим матрицу эмбеддингов слов
word_vectors = (U @ np.diag(s))[:, :k]

# считаем эмбеддинги документов и транспонируем результат,
# чтобы обеспечить совместимость с пространством слов
doc_vectors = (np.diag(s) @ VT)[:k].T
```

```python
word_vectors.shape, doc_vectors.shape
```

```python
# создадим новый документ со словом,
# для которого необходимо найти наиболее близкий документ
new_doc = np.array(['вкусный'])
```

```python
# создаем вектор нового документа
search_index = np.where(words == new_doc)[0]
search_index
```

```python
# выведем значения списка words
words
```

```python
# находим вектор нового документа
search_vector = np.zeros((1, k)) + word_vectors[search_index]
search_vector
```

```python
# Функция возвращает индекс ближайшего к `search_vector` вектора из строк `vectors`.
def find_nearest_vector_id(search_vector, vectors):
		# L2 расстояние между векторами.
    distances = np.mean((search_vector - vectors) ** 2, axis=1)
		# Индексы векторов с наименьшим расстоянием.
    return np.argmin(distances)
```

```python
# найдем индекс ближайшего к новому веткору документа
index = find_nearest_vector_id(search_vector, doc_vectors)
```

```python
# выведем значение индекса
index
```

```python
# выведем значение документа, наиболее близкого к новому индексу
print(preprocessed_texts[index])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

price = [2550, 1920, 2200, 1950, 1987, 2030, 2655, 1855]
age = [1, 2, 1, 2, 2, 2, 0, 2]
km = [40, 39269, 36822, 32834, 35255, 59005, 10910, 20627]

labs = ['Цена', 'Возраст', 'Пробег']

print("Матрица ковариации:")
def create_martix(data):
    # выделяем каждую случайную величину
    # в отдельную переменную
    x, y, z = data[:, 0], data[:, 1], data[:, 2]

    # вычислим выборочное среднее каждой случайной величины
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    z_mean = np.mean(z)

    # для вычисления выборочной ковариации вычислим длину массива - 1
    len_array = len(data) - 1

    # вычисляем выборочные дисперсии случайных величин
    var_x = np.var(x, ddof = 1)
    var_y = np.var(y, ddof = 1)
    var_z = np.var(z, ddof = 1)

    # вычисляем выборочные ковариации
    cov_xy = ((x - x_mean) @ (y - y_mean)) / len_array
    cov_xz = ((x - x_mean) @ (z - z_mean)) / len_array
    cov_yz = ((y - y_mean) @ (z - z_mean)) / len_array

    # составляем матрицу ковариации
    matrix = np.array([
        [var_x, cov_xy, cov_xz],
        [cov_xy, var_y, cov_yz],
        [cov_xz, cov_yz, var_z]
    ])

    return matrix

# сформируем массив из трех признаков
data = np.array([price, age, km]).T
print(data)

# получим матрицу ковариации из функции
cov_mat = create_martix(data)
print(cov_mat)

sns.heatmap(
	cov_mat, # не забудьте определить переменную
	annot=True,
	fmt='g',
	xticklabels=labs,
	yticklabels=labs,
    cmap = 'RdBu',
    linewidth = .9
)
plt.show()
```

## **Урок 3. Применения метода главных компонент**

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
        [5, 12],
        [2, 5],
        [0, 2],
        [2, 10],
        [6, 25],
        [3, 14],
        [2, 12],
        [3, 23],
        [7, 26],
        [2, 16]
])

# Создаём объект PCA
pca = PCA(n_components=1)

# Сжимаем данные.
data_compressed = pca.fit_transform(data)

# Строим собственный вектор, на который проецировались данные
# Последний множитель помогает в красивой отрисовке
magnificent_coef = 3
principal_vector = pca.components_[0] * np.sqrt(pca.explained_variance_) * magnificent_coef

# Рассчитываем координаты точек для визуализации главной компоненты
x_min = pca.mean_[0] - principal_vector[0]
x_max = pca.mean_[0] + principal_vector[0]

y_min = pca.mean_[1] - principal_vector[1]
y_max = pca.mean_[1] + principal_vector[1]

plt.figure(figsize=(18, 9))

cmap = "viridis" # Задаём цветовую схему

# Строим график с исходными данными
plt.subplot(121)
# Исходные данные.
plt.scatter(data[:, 0], data[:, 1], c=data_compressed.flatten(), cmap=cmap, s=100)
# Соответствующие точки на оси
points_on_line = pca.inverse_transform(data_compressed)
# Отрезки между точками и осью
plt.plot(
    np.stack([data[:, 0], points_on_line[:, 0]], axis=1).T,
    np.stack([data[:, 1], points_on_line[:, 1]], axis=1).T,
    zorder=-2,
    linewidth=2,
    c="black",
)
# Визуализируем главную ось
plt.plot(
    [x_min, x_max],
    [y_min, y_max],
    linewidth=3,
    zorder=-2,  # Размещаем линию под точками
    label="Ось главной компоненты",
)
# Задаём формат графика.
plt.xlim(-10, 15)
plt.ylim(-2, 30)
# Делаем отсечки на осях одинаковыми по масштабу
plt.gca().set_aspect("equal")
plt.legend()
plt.title("Исходные данные")

# Строим график со сжатыми данными
plt.subplot(122)
# Трансформированные данные
plt.scatter(
    data_compressed,
    np.zeros(data_compressed.shape),
    c=data_compressed,
    cmap=cmap,
    s=100,
)
# Главная ось.
plt.plot(
    [-20, 20],
    [0, 0],
    zorder=-2,  # Размещаем линию под точками
    linewidth=2,
    label="Ось главной компоненты",
)
# Задаём формат графика.
plt.xlim(-15, 15)
plt.ylim(-10, 10)
# Делаем отсечки на осях одинаковыми по масштабу
plt.gca().set_aspect("equal")
plt.yticks([])  # Отключаем засечки на оси Oy
plt.title("Трансформированные данные")

plt.legend()
```

**Применение PCA для улучшения качества работы линейной регрессии**

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X, y = load_diabetes(return_X_y = True)

# random_state необходимо зафиксировать,
# чтобы при каждом запуске разбиение было одинаковым.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = Lasso(alpha = .01).fit(X_train, y_train)

error = mean_absolute_error(y_test, lr.predict(X_test))
print(f"Средняя ошибка линейной регрессии: {error:.4f} баллов")

const_error = mean_absolute_error(y_test, np.ones(len(y_test)) * y_train.mean())
print(f"Средняя ошибка константного предикта: {const_error:.4f} баллов")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data_url = 'https://code.s3.yandex.net/Math/datasets/credit_rus.csv'
df = pd.read_csv(data_url)
print(df.columns)
X = df.drop(['Кредитный лимит, $'], axis=1).values
y = df['Кредитный лимит, $'].values

# random_state необходимо зафиксировать,
# чтобы при каждом запуске разбиение было одинаковым.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# нормализуем обучающие данные
X_train_normalized = X_train / X_train.std(axis = 0)
X_test_normalized = X_test / X_train.std(axis = 0)

# подбираем число главных компонент
# объявляем объект PCA
n_cols = X.shape[1]
pca = PCA(
    n_components = n_cols,
    random_state = 42).fit(X_train_normalized)

# выведем сингулярные значения
print(pca.singular_values_)

# визуализируем сингулярные значения
plt.bar(range(1, X.shape[1] + 1), pca.singular_values_)

# вновь объявляем объект PCA
# с подобранным числом главных компонент
n_components = 8
pca = PCA(
    n_components = n_components,
    random_state = 42)

# сжимаем обучающие данные
X_train_compressed = pca.fit_transform(X_train_normalized)
X_test_compressed = pca.transform(X_test_normalized)

# объявляем модель
model = LinearRegression()

# обучаем модель
model.fit(X_train_compressed, y_train)

# получаем предсказания
predictions = model.predict(X_test_compressed)

# вычислим MAE
error = mean_absolute_error(y_test, predictions)

print(n_components)
print(error)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

data_url = 'https://code.s3.yandex.net/Math/datasets/phone_prices_ru.csv'
data = pd.read_csv(data_url)
print(data.columns)
X = data.iloc[:, 2:-1].values
y = data.iloc[:, -1].values

# random_state необходимо зафиксировать,
# чтобы при каждом запуске разбиение было одинаковым.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# нормализуем данные
x_train_normalized = X_train / X_train.std(axis = 0)
x_test_normalized = X_test / X_train.std(axis = 0)

# определяем сингулярные значения
n_components = x_train_normalized.shape[1]

# объявляем объект PCA - для предварительного отбора
# сингулярных значений
pca = PCA(n_components = n_components,
          random_state = 42)

# применяем PCA к датасету X_train_normalized
pca.fit(x_train_normalized)

# выведем величины сингулярных значений
print(pca.singular_values_)

# визуализируем сингулярные значения
plt.bar(
    range(1, x_train_normalized.shape[1] + 1),
    pca.singular_values_)
plt.show()

# число отобранных для ML главных компонент
n_components = 19

# объявляем объект PCA - для сжатия данных
# для оптимизации результатов машинного обучения
pca = PCA(
    n_components = n_components,
    random_state = 42
)

# сжимаем данные
x_train_compressed = pca.fit_transform(x_train_normalized)
x_test_compressed = pca.transform(x_test_normalized)

# объявляем модель ML
model_lr = LinearRegression()

# обучаем модель
model_lr.fit(x_train_compressed, y_train)

# получаем предсказания
predictions = model_lr.predict(x_test_compressed)

# вычисляем MAE
error = mean_absolute_error(
    y_test,
    predictions
)

print(n_components)
print(error)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import requests
import io

data_url = 'https://code.s3.yandex.net/Math/datasets/lfw_people.npz'
response = requests.get(data_url)
lfw_people = np.load(io.BytesIO(response.content))

X, y = lfw_people['X'].astype(int), lfw_people['y'].astype(int)

# центрируем данные
X_centered = X - np.mean(X, axis = 0)

# нормализуем данные
X_normalised = X_centered / X_centered.std(axis = 0)

# сжимаем данные с помощью TSNE
X_tsne = TSNE(
    n_components = 2,
    random_state = 42
).fit_transform(X_normalised)

# визуализируем сжатые данные
plt.scatter(
    x = X_tsne[:, 0],
    y = X_tsne[:, 1],
    c = y
)
plt.show()

def plot_images(X_tsne):
    fig, ax = plt.subplots(figsize=(24, 24))
    plt.xlim(X_tsne[:, 0].min() - 5, X_tsne[:, 0].max() + 5)
    plt.ylim(X_tsne[:, 1].min() - 5, X_tsne[:, 1].max() + 5)
    for (x0, y0), x, label in zip(X_tsne, lfw_people['images'].astype(np.uint8), y):
        ab = AnnotationBbox(
            OffsetImage(x),
            (x0 + np.random.randn() * 2, y0 + np.random.randn() * 2),
            frameon=True,
            pad=0.01,
            bboxprops =dict(edgecolor=['red', 'blue'][label]),
        )
        ax.add_artist(ab)
    plt.show()

plot_images(X_tsne)
```

```python
# усредненное количество пользователей
n1_hist = 2_621_000
n2_hist = 2_645_900

# усредненная дисперсия конверсии
# в оформление подписки
var1_hist = 0.041
var2_hist = 0.048

# сумма значений для ошибок 1-го и 2-го рода z_0025 + z_02
z_summ = -2.802

# усредненные значения конверсии в оформление подписки
CR1_hist = 0.042
CR2_hist = 0.051

# желаемый прирост метрики в %
purpose = 1.5
```

```python
# MDE для 1-й недели
mde_1 = - z_summ * ((4 * var1_hist / n1_hist) ** 0.5)

# MDE для 2-й недели
mde_2 = - z_summ * ((4 * var2_hist / n2_hist) ** 0.5)

print(f'mde_1: {mde_1}')
print(f'mde_2: {mde_2}')
```

```python
# относительный MDE для 1-й недели
mde1_ratio = (mde_1 / CR1_hist) * 100

# относительный MDE для 2-й недели
mde2_ratio = (mde_2 / CR2_hist) * 100

print(f'mde1_ratio: {mde1_ratio:.3}')
print(f'mde2_ratio: {mde2_ratio:.3}')
```

# **A/A-тест. Расчет p-value для конверсии в оплату любой подписки**

```python
# конверсия в оплату любой подписки
cr_test = 0.1526
cr_control = 0.1524

# размер группы (выборки)
n_test = 7_863_369
n_control = 7_864_235
```

```python
# рассчитаем дисперсию в тестовой и контрольной группах
var_test = cr_test * (1 - cr_test)
var_control = cr_control * (1 - cr_control)
```

```python
# рассчитаем z-статистику в рамках z-теста для пропорций
part1 = cr_test - cr_control
part2 = var_test / n_test
part3 = var_control / n_control

z = part1 / (part2 + part3) ** 0.5
z
```

**p_value = 0.271**

Не отвергаем нулевую гипотезу. Разбиение валидно.

**По итогам А/В-теста С помощью Т-теста проверим, что метрика не просто выросла, а выросла не меньше, чем на желаемые 1.5%**

```python
# значение метрики конверсии в тестовой и контрольной группах
p_test = 0.0526
p_control = 0.0512

# прогнозируемый абсолютный прирост
diff = 0.000765

# размер тестовой и контрольной группы
n_test = 1_322_630
n_control = 1_321_882

# дисперсия в тестовой и контрольной группах
var_test = p_test * (1 - p_test)
var_control = p_control * (1 - p_control)
```

```python
# рассчитаем z-статистику в рамках z-теста для пропорций
part1 = (p_test - p_control) - diff
part2 = var_test / n_test
part3 = var_control / n_control

z = part1 / (part2 + part3) ** 0.5
z
```

**p-value = 0.02**

Не отвергаем нулевую гипотезу.
